```python
import numpy as np
from sklearn.linear_model import LogisticRegression 
from sklearn.linear_model import LogisticRegressionCV
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier  
from sklearn import metrics
```

```python
train, test = train_test_split(df, test_size = 0.3)
print(train.shape)
print(test.shape)
```

```python
train_X=train.drop('diagnosis',axis=1)
train_y=train.diagnosis
test_X=test.drop('diagnosis',axis=1)
test_y=test.diagnosis
```

```python
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(train_y.shape)
```

```python
model=LogisticRegressionCV(cv=20)
model.fit(train_X,train_y)
model.score(test_X,test_y)
```

```python
model = LogisticRegression()
model.fit(train_X,train_y)
prediction=model.predict(test_X)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,test_y)*100 ,'percent')
```

```python
model=KNeighborsClassifier(n_neighbors=3) 
model.fit(train_X,train_y)
prediction=model.predict(test_X)
print('The accuracy of the KNN is',metrics.accuracy_score(prediction,test_y)*100, 'percent')
```

```python
features_mean= list(df.columns[1:11])
```

```python
corr = df[features_mean].corr()
corr.shape
```

```python
plt.figure(figsize=(14,14))
sns.heatmap(corr, cbar=True, square= True, fmt='.2f',annot=True,annot_kws={'size':15}, cmap='Greens')
# Observations:
# radius, perimeter, area are correlated (obviously). We consider parameter because it's most correlated to other features.
# compactness_mean, concavity_mean, concave_points_mean are correlated .
# smoothness_mean, symmetry_mean and fractal_dimension_men are barely correlated to anything else

# So finally we select the parameters perimeter,texture,compactness,symmetry
```

```python
prediction_variables = ['perimeter_mean','smoothness_mean','symmetry_mean','compactness_mean']
corr2=df[prediction_variables].corr()
plt.figure(figsize=(14,14))
sns.heatmap(corr2, cbar=True, square= True, fmt='.2f',annot=True,annot_kws={'size':15},cmap="Greens")
```

```python
train, test = train_test_split(df, test_size = 0.3)
print(train.shape)
print(test.shape)
```

```python
train_X=train[prediction_variables]
train_y=train.diagnosis
test_X=test[prediction_variables]
test_y=test.diagnosis
```

```python
model = LogisticRegression()
model.fit(train_X,train_y)
prediction=model.predict(test_X)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,test_y)*100 ,'percent')
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=50)
model.fit(train_X,train_y)
prediction=model.predict(test_X)
print('The accuracy with Random Forest is',metrics.accuracy_score(prediction,test_y)*100)
```
